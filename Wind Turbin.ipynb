```python
x= dataset.iloc[:,:2]
```

```python
y= dataset.iloc[:,-1:]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
import keras #Building neural network model back end it has tensorflow
from keras.models import Sequential # initializing model
from keras.layers import Dense # used for adding the neural network layers
```

```python
model=Sequential()#instance
```

```python
model.add(Dense(input_dim=2,init="random_uniform",activation="tanh",output_dim=1)) #input_dim=no of neurons to have in the input layer(no of independent variables),init=adding weights,activation function,output_dim=next hidden layer
#best thing for output_dim is input_dim/2
```

```python
model.add(Dense(output_dim=5,init="random_uniform",activation="relu"))# adding hidden layer similary like adding weights
```

```python
model.add(Dense(output_dim=1,init="random_uniform",activation="tanh"))#adding output layer
```

```python
model.compile(optimizer='adam',loss='mse',metrics=['mse'])#adam=batch gradent descent
```

```python
model.fit(x_train,y_train,epochs=1000,batch_size=5)#epochs no of iteration 
```

```python
model.save("wind.h5")
```
